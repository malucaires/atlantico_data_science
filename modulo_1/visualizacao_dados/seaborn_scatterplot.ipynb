```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
diamonds = sns.load_dataset('diamonds')

diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds = diamonds [
    diamonds.cut.isin(['Premium', 'Good']) &
    diamonds.color.isin(['D', 'F', 'J'])
    ].sample(n=100, random_state=22)

diamonds.shape
```

```python
sns.set_style('dark')
```

```python
sns.scatterplot(x='carat', y= 'price',
               data = diamonds)
```

```python
sns.scatterplot(x='x', y= 'y',
               data=diamonds)
```

```python
sns.scatterplot(x='carat', y= 'price', hue='cut',
                data = diamonds)
```

```python
sns.scatterplot(x='carat', y= 'price', hue='depth',
                data = diamonds)
```

```python
sns.scatterplot(x='carat', y='price', s=100,
                style='cut',
                data=diamonds)
```

```python
sns.scatterplot(x='carat', y='price', size='cut',
                data=diamonds)
```

```python
sns.scatterplot(x='carat', y='price',
                data=diamonds,
                s=200,
                marker='*',
                edgecolor='black'
               )
```

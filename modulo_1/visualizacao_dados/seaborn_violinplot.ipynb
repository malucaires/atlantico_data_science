```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement)
```

```python
sns.violinplot(x='cylinders', y='displacement',
               data=cars
              )
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars
              )
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan','europe'])]
              )
plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan','europe'])],
               split=True
              )
plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan','europe'])],
               split=True,
               inner='quartilies'
              )
plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan','europe'])],
               split=True,
               scale='count'
              )
plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan','europe'])],
               split=True,
               scale='count',
               scale_hue=False
              )
plt.legend(loc=2)
```

```python
(
    cars[cars.origin.isin(['japan', 'europe'])]
    .groupby('cylinders')
    .origin
    .value_counts()
)
```

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
               order=['japan', 'europe', 'usa']
              )
```

```python
sns.violinplot(x='origin', y='displacement',
               data = cars,
               linewidth=3
              )
```

```python
sns.violinplot(x='origin', y='displacement',
               data = cars,
               bw=0.4
              )
```

```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('penguins')
df.head()
```

```python
len(df)
```

```python
#One-hot Encoding é uma representação vetorial em que todos os elementos em um vetor são 0, exceto um, que tem 1 como valor, onde 1 representa um boolean especificando uma categoria do elemento. [1,0,0]; [0,1,0]; e [0,0,1]

pd.get_dummies(df[['island']], prefix=['island'])
```

```python
#Dummy enconding é bem parecido com One-hot enconding. A única diferença é que, neste caso, utilizamos um valor binário a menos, colocando zero nas duas variáveis quando elas se referiam ao valor da outra variável.[0,0],[1,0][0,1]

pd.get_dummies(df[['island']], prefix=['island'], drop_first=True)
```
